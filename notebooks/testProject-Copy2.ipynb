```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dfTrain = pd.read_csv("train.csv",header=0)
dfTest = pd.read_csv("test.csv",header=0)
yTest = pd.read_csv("sample_submission.csv",header=0)
```

```python
dfTrain.head(2)
```

```python
dfTrain = dfTrain.set_index("Id")
dfTest = dfTest.set_index("Id")
yTest = yTest.set_index("Id")
dfTrain.head()
```

```python
for col in dfTrain.columns[dfTrain.isna().any()].tolist():
    if(dfTrain[col].dtype == "object"):
        dfTrain[col].fillna("unknown", inplace=True)
    if((dfTrain[col].dtype == "float64") or (dfTrain[col].dtype == "int64")):
        dfTrain[col].fillna(dfTrain[col].mean(), inplace=True)
for col in dfTest.columns[dfTest.isna().any()].tolist():
    if(dfTest[col].dtype == "object"):
        dfTest[col].fillna("unknown", inplace=True)
    if((dfTest[col].dtype == "float64") or (dfTest[col].dtype == "int64")):
        dfTest[col].fillna(dfTest[col].mean(), inplace=True)
```

```python
dfTrain.head()
```

```python
XTrain = dfTrain.drop(['SalePrice'],1)
XTrain = XTrain.select_dtypes(exclude=['object'])
yTrain = np.array(dfTrain['SalePrice'])
XTest = dfTest.select_dtypes(exclude=['object'])
yTest = np.array(yTest['SalePrice'])
```

```python
XTrain.head()
```

```python
from sklearn.preprocessing import scale
XTrainScaled = scale(XTrain)
XTestScaled = scale(XTest)
# XTrainScaled = np.array(XTrain)
# XTestScaled = np.array(XTest)
```

```python
from sklearn import svm
from sklearn.linear_model import LinearRegression
clf = svm.SVR(C=10000,epsilon=5,gamma=0.0005)
clf.fit(XTrainScaled,yTrain)
```

```python
from sklearn.metrics import accuracy_score
yPrdeict = clf.predict(XTestScaled)
r2 = clf.score(XTest,yPrdeict)
r2_train = clf.score(XTrain,yTrain)

# for i in range(50):
#     print("ytest:{}  ypredict:{}".format(yTest[i],yPrdeict[i]))
r2_train
```
